```python

import pandas as pd

file_path = "/content/API_SP.POP.TOTL_DS2_en_csv_v2_900.csv"


data = pd.read_csv(file_path, skiprows=4)


print("Dataset loaded successfully!")
print(data.head())


year = "2022"
if year in data.columns:
    population_data = data[["Country Name", year]].dropna()
    population_data[year] = population_data[year].astype(float)


    import matplotlib.pyplot as plt

    plt.figure(figsize=(10, 6))
    plt.hist(population_data[year], bins=30, color='blue', edgecolor='black')
    plt.title(f"Population Distribution in {year}", fontsize=16)
    plt.xlabel("Population", fontsize=14)
    plt.ylabel("Number of Countries", fontsize=14)
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.show()


    top_10 = population_data.nlargest(10, year)

    plt.figure(figsize=(10, 6))
    plt.barh(top_10["Country Name"], top_10[year], color='green')
    plt.title(f"Top 10 Most Populated Countries in {year}", fontsize=16)
    plt.xlabel("Population", fontsize=14)
    plt.ylabel("Country", fontsize=14)
    plt.gca().invert_yaxis()  # Invert Y-axis for better readability
    plt.show()
else:
    print(f"The year {year} is not available in the dataset.")
```
